```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
normal_df = pd.read_csv("Data/NormalMode.csv")
attack1_df = pd.read_csv("Data/Attack1_flood80.csv")
attack2_df = pd.read_csv("Data/Attack2_flood80.csv")
attack3_df = pd.read_csv("Data/Attack3_Ping_and_attack.csv")
attack_df = pd.concat([attack1_df, attack2_df,attack3_df], axis=0)
test_data = pd.read_csv("Data/two consecutive attacks_filter.csv")

normal_df["attack"] = 0
attack_df["attack"] = 1
combined_df = pd.concat([normal_df, attack_df], axis=0)
```

```python
combined_df.head()
```

```python
combined_df.drop(columns=["timestamp"]).corr()
```

```python
# Select relevant features for modeling
selected_features = ['responsetime', 'freeheapmemory', 'powerconsumption']

X = combined_df[selected_features]
y = combined_df['attack']

standard_scaler = StandardScaler()
X_scaled = standard_scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Train a simple logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python

# Make predictions on the test set
y_pred = logreg.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_report_output = classification_report(y_test, y_pred)
```

```python
# Print results
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:\n", classification_report_output)
```
